```python
away_team_df = pd.read_csv("2023_away_teams_stats.csv")
away_team_df
```

```python
home_team_df = pd.read_csv("2023_home_teams_stats.csv")
home_team_df
```

```python
# Team specific analysis (Team="arsenal")
arsenal_home_df = home_team_df.groupby('Home team name').get_group('Arsenal')
arsenal_away_df = away_team_df.groupby('away team name').get_group('Arsenal')

# Calculate average home possession of Arsenal
home_team_df_poss = home_team_df['Ball Possession'].str.replace('%', '').astype(float)
average_home_possession = home_team_df_poss.mean()

# Calculate average away possession of Arsenal
away_team_df_poss = away_team_df['Ball Possession'].str.replace('%', '').astype(float)
average_away_possession = away_team_df_poss.mean()

average_possessions = [average_home_possession, average_away_possession]
labels = ['Home Possession', 'Away Possession']

plt.figure(figsize=(4, 4))
plt.pie(average_possessions, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Arsenal Average Home vs. Away Possession')
plt.axis('equal')  
plt.show()
```

```python
stats = ['Shots on Goal', 'Shots off Goal', 'Total Shots', 'Blocked Shots', 'Shots insidebox', 'Shots outsidebox']

arsenal_home_stats = arsenal_home_df[arsenal_home_df['Home team name'] == 'Arsenal'][stats].mean()
arsenal_away_stats = arsenal_away_df[arsenal_away_df['away team name'] == 'Arsenal'][stats].mean()

num_stats = len(stats)
angles = np.linspace(0, 2 * np.pi, num_stats, endpoint=False).tolist()

arsenal_home_stats = np.concatenate((arsenal_home_stats, [arsenal_home_stats[0]]))
arsenal_away_stats = np.concatenate((arsenal_away_stats, [arsenal_away_stats[0]]))
angles += angles[:1]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), subplot_kw=dict(polar=True))

ax1.fill(angles, arsenal_home_stats, color='red', alpha=0.25)
ax1.plot(angles, arsenal_home_stats, color='red', linewidth=2)

ax1.fill(angles, arsenal_home_stats, color='red', alpha=0.25)

ax1.set_yticklabels([])
ax1.set_xticks(angles[:-1])
ax1.set_xticklabels(stats, fontsize=10, fontweight='bold', rotation=45)

ax1.set_title('Arsenal Home Match Statistics', size=20, color='red')

ax2.fill(angles, arsenal_away_stats, color='blue', alpha=0.25)
ax2.plot(angles, arsenal_away_stats, color='blue', linewidth=2)

ax2.fill(angles, arsenal_away_stats, color='blue', alpha=0.25)

ax2.set_yticklabels([])
ax2.set_xticks(angles[:-1])
ax2.set_xticklabels(stats, fontsize=10, fontweight='bold', rotation=45)

ax2.set_title('Arsenal Away Match Statistics', size=20, color='blue')

plt.tight_layout()

plt.show()
```
